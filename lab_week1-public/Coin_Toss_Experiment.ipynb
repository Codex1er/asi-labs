# Advanced Statistical Inference -- Bayesian inference for a coin tossing experiment


The beta probability density function is defined as:

$$
p(r \vert \alpha, \beta) = \dfrac{\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)}r^{\alpha-1}(1-r)^{\beta-1} 
$$

where $0 \le r \le 1$ and $\alpha, \beta> 0$.
You can use the `plot_beta()` function.


```python
from scipy.stats import beta
from scipy.special import gamma as gamma_func, comb
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
```

```python
def plot_beta(a,b, plot=False):
    plt.figure(figsize=(15,5))
    x = np.linspace(0, 1, 1000)
    plt.plot(x, beta.pdf(x, a, b),'r')
    plt.title("beta pdf: alpha=" + str(a) + " beta=" + str(b))
    plt.show()
```

```python
plot_beta(a=2, b=2)
```

```python
def coin1(N):
    return sum(np.random.random(N)<=0.3)
def coin2(N):
    return sum(np.random.random(N)<=0.8)
```

```python
yN = [coin2(N) for N in range(1000)]
yN1 = [coin1(N) for N in range(1000)]
plt.figure(figsize=(10,5))
plt.plot(list(range(1000)), yN)
plt.plot(list(range(1000)), yN1)
plt.xlabel("Number of tosses")
plt.ylabel("Number of heads")

plt.title("coin2")
plt.grid()
plt.show()
```

We can see the linear results we expect from a binomial law, In blue, the coin2 with a ramp of 0.8, what is the probability of this random variable to give a 1, and in orange the same for coin1, with a probability of 0.3. The noise of theses linear functions can be attribuated to the randomness of the binomial variable. We can prove that the noise expectancy growth linearly with the number of tosses.

So, `coin2()` is biased to generate more heads (roughly 8 out of 10).

```python
N = 1000

yN1, yN2 = coin1(N), coin2(N)

# Assume coins are fair

alpha1 = 50
beta1 = 50




# Compute the posterior 'gamma' and 'delta' parameters
gamma1, gamma2 = alpha1+yN1, alpha1+yN2 # TO COMPLETE
delta1, delta2 = beta1 + (N-yN1), beta1 + (N-yN2) # TO COMPLETE

plt.figure(figsize=(15,5))
x = np.linspace(0, 1, 1000)
plt.plot(x, beta.pdf(x, alpha1, beta1),'r', label='prior')
plt.plot(x, beta.pdf(x, gamma1, delta1),'b', label='posterior coin1')
plt.plot(x, beta.pdf(x, gamma2, delta2),'g', label='posterior coin2')
plt.legend()
plt.grid()
plt.show()
```

What we can see here is that no matter what the prior is, the bayesian update make the posterior be true. The posterior just may require more that to be accurate if the prior is very different.

```python
N = [1, 5, 10, 100, 1000]
cs = ['b', 'c', 'g', 'y', 'r']
a = 2
b = 2
x = np.linspace(0, 1, 100)
plt.figure(figsize=(15,8))
for i,n in enumerate(N):
    # Compute the posterior distribution for coin2 given 'n' tosses 
    yN1 = coin2(n) # TO COMPLETE
    gamma = a + yN1
    delta = b+(n-yN1)
    plt.plot(x, beta.pdf( x, gamma, delta ),color=cs[i], label="N="+str(n)) # TO COMPLETE
    
plt.plot(x, beta.pdf( x, a, b ),'k+', label='prior') # TO COMPLETE
plt.title("Coin2")
plt.legend()
plt.grid()
plt.show()
```

We can see that for a limited amount of data, the prior influence a lot the prosteriors. Moreover, the curve with an high quantity of data are more acute than other, showing that the flateness of the posterior curves somehow shows the accuracy we have using this model.
